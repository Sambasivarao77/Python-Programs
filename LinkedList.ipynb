Problems' with normal list  

1.   If we have to insert a element at positions 1 in the list which already has few elements then we need to swap all elements succeding elements by one position.  Time complexity to insert is O(N)
2.   While iserting a new element if the contiguous memroy if full then we have to copy all the list elemnts to a new meory where we have more space. 

These two problems can be fixed with linked list
